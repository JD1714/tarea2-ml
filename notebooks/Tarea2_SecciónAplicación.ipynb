## Cambios en la aplicación

Para los cambios de la aplicación los separaremos cada uno en un encabezado:

### Mapeo de los Operadores

Modificamos la aplicación web, sustituyendo el antiguo sistema para identificar cada operador (del 0 al 5), a mapearlos con su respectivo código ASCII. Esto tuvimos que hacerlo en todos los lugares donde se usaba la configuración anterior, es decir, en el archivo game_state.py y jugar.py.

### Preprocesamiento de las imágenes de dígitos

Siguiendo lo hecho en el notebook, modificamos las funciones para procesar las imágenes (prepare_digit_image) dentro de la aplicación, para que solamente usara la función proporcionada transform_image_to_mnist.

### Trazos de los Operadores

A su vez, para que ambos tipos de imágenes tuvieran el mismo preprocesamiento y por simplicidad también, entrené mis modelos de operadores con trazos claros y fondos oscuros. Por lo que, en la aplicación tuve que modificar la función de prepare_operator_image para que usara transform_image_to_mnist.
